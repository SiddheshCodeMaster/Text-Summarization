```python
def remove_emojis(text):
  emoji_pattern = re.compile("["
          u"\U0001F600-\U0001F64F"  # emoticons
          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
          u"\U0001F680-\U0001F6FF"  # transport & map symbols
          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', text)
```

```python
df_copy['clean_abstract'] = df_copy['clean_abstract'].apply(remove_emojis)
```

#### T5-Large-Xsum-CNN:

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer

def t5_large_summarizer(text):
  checkpoint = "sysresearch101/t5-large-finetuned-xsum-cnn"
  tokenizer = AutoTokenizer.from_pretrained(checkpoint)
  model = model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
  ARTICLE_TO_SUMMARIZE = text
  # generate summary

  input_ids = tokenizer.encode(ARTICLE_TO_SUMMARIZE, return_tensors='pt')
  summary_ids = model.generate(input_ids,
            min_length=50,
            max_length=400,
            num_beams=10,
            repetition_penalty=2.5,
            length_penalty=1.0,
            early_stopping=True,
            no_repeat_ngram_size=2,
            use_cache=True,
            do_sample = True,
            temperature = 0.8,
            top_k = 50,
            top_p = 0.95)

  summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
  return summary_text
```

```python
df_copy['Summarized_by_T5'] = df_copy['clean_abstract'].apply(t5_large_summarizer)
df_copy.head()
```

```python
output = df_copy['Summarized_by_T5'][0]
print(output)
```

```python
from nltk.translate.bleu_score import sentence_bleu

ref = output.split()

print('BLEU Score for the first record is {}'.format(sentence_bleu(ref, df_copy['Summary'][0].split())))
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
scores = scorer.score(df_copy['Summary'][0],t5_large_summarizer(df_copy['clean_abstract'][0]))
print(scores)
```

#### Bert small Bert 2 bert model: 

```python
from transformers import BertTokenizerFast, EncoderDecoderModel
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
checkpoint = "mrm8488/bert-small2bert-small-finetuned-cnn_daily_mail-summarization"
def bert_smallbert2bert(text):
  tokenizer = BertTokenizerFast.from_pretrained(checkpoint)
  model = EncoderDecoderModel.from_pretrained(checkpoint).to(device)
  # cut off at BERT max length 512
  inputs = tokenizer([text], padding="max_length", truncation=True, max_length=512, return_tensors="pt")
  input_ids = inputs.input_ids.to(device)
  attention_mask = inputs.attention_mask.to(device)

  output = model.generate(input_ids, attention_mask=attention_mask)

  return tokenizer.decode(output[0], skip_special_tokens=True)
```
